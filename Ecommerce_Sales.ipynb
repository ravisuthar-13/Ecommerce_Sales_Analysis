```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
```

```python
data.head(2)
```

```python
# 1. Total Sales and Profit by Region

region_stats = data.groupby('region')[['sales', 'profit']].sum().sort_values(by='sales', ascending=False)
print(region_stats)
```

```python
# 2. Top Performing Categories

top_categories = data.groupby('category')[['profit']].sum().sort_values(by='profit', ascending=False).head(5)
print(top_categories)
```

```python
top_cities = data.groupby('city')['sales'].sum().sort_values(ascending=False).head(5)
print(top_cities)
```

```python
# 4. Correlation between Discount and Profit

discount_profit_corr = data['discount'].corr(data['profit'])
print(discount_profit_corr)
```

```python
# 5. Preferred Payment Modes

payment_counts = data['payment_mode'].value_counts()
print(payment_counts)
```

```python
# 6. Top 5 Cities by Sales

top_cities = data.groupby('city')['sales'].sum().sort_values(ascending=False).head(5)
print(top_cities)
```
